```python
data2=pd.read_csv("netflix_titles.csv")
print(data2)
```

```python
print(data2.head())
print(data2.info())
```

```python
print(data2.isna().sum())
```

```python
print(data2.duplicated().sum())
data2 = data2.drop_duplicates()
```

```python
for col in ['director', 'cast', 'country', 'rating']:
    if col in data2.columns:
        data2[col] = data2[col].fillna(data2[col].mode()[0])
```

```python
if 'date_added' in data2.columns:
    data2['date_added'] = pd.to_datetime(data2['date_added'], errors='coerce')
    data2['date_added'] = data2['date_added'].fillna(data2['date_added'].mode()[0])
```

```python
data2.columns = data2.columns.str.lower()
data2.columns = data2.columns.str.replace(" ", "_")
```

```python
print(data2.isna().sum())
print(data2.info())
```

```python
data1.to_csv("iris_cleaned.csv", index=False)
data2.to_csv("netflix_cleaned.csv", index=False)
```
